```python
h11 = []
for i in range(4):
    h11.append(p1[i][0])
for i in range(4):
    h11.append(0)
# print h11

h12 = []
for i in range(4):
    h12.append(p1[i][1])
for i in range(4):
    h12.append(0)
# print h12

h13 = []
for i in range(4):
    h13.append(1)
for i in range(4):
    h13.append(0)
# print h13

h21 = []
for i in range(4):
    h21.append(0)
for i in range(4):
    h21.append(p1[i][0])
# print h21

h22 = []
for i in range(4):
    h22.append(0)
for i in range(4):
    h22.append(p1[i][1])
# print h22

h23 = []
for i in range(4):
    h23.append(0)
for i in range(4):
    h23.append(1)
# print h23

h31 = []
for i in range(4):
    h31.append(-1*p1[i][0]*p2[i][0])
for i in range(4):
    h31.append(-1*p1[i][0]*p2[i][1])
# print h31

h32 = []
for i in range(4):
    h32.append(-1*p1[i][1]*p2[i][0])
for i in range(4):
    h32.append(-1*p1[i][1]*p2[i][1])
# print h32

a = np.array([h11,h12,h13,h21,h22,h23,h31,h32])
b = []
for i in range(4):
    b.append(p2[i][0])
for i in range(4):
    b.append(p2[i][1])
b = np.array(b)

A = np.zeros((8,9))
A[:,0] = h11
A[:,1] = h12
A[:,2] = h13
A[:,3] = h21
A[:,4] = h22
A[:,5] = h23
A[:,6] = h31
A[:,7] = h32
A[:,8] = -b
# print A

A = Matrix(A)
H = A.nullspace()[0]
H = np.array(H)
H = H.reshape((3,3))
# print H
```

```python
newimage = np.zeros(image.shape)
for x in range(image.shape[1]):
    for y in range(image.shape[0]):
        res = np.dot(H,np.array([x,y,1]))
        x1 = int(round(res[0]/res[2]))
        y1 = int(round(res[1]/res[2]))
        if x1>=0 and x1<image.shape[1]:
            if y1>=0 and y1<image.shape[0]:
                newimage[y1,x1,:] = image[y,x,:]

plt.figure(figsize=(10,10))
plt.imshow(newimage)
# plt.imshow(image)
```
